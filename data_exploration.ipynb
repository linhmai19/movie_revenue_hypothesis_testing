```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#import the rt.movie_info.cvs file for comparation between
#theater_date vs box_office
#genre vs box_office
#runtime vs box_office
rt_movie_info = pd.read_csv('data/rt.movie_info.csv')
rt_movie_info.head()
```

```python
#eliminate the first column and set 'id' column as index
rt_movie_info.set_index('id', inplace=True)
```

```python
rt_movie_info.head()
```

```python
#check the number of missing values in each column
rt_movie_info.isna().sum()
```

```python
#change the type of 'theater_date' column from object to datetime 
rt_movie_info['theater_date'] = pd.to_datetime(rt_movie_info['theater_date'])
print(rt_movie_info['theater_date'].dtype)
```

```python
rt_movie_info
```

```python
movies_from_2000_2018 = rt_movie_info[(rt_movie_info['theater_date'] > '2000-01-01') 
                                      & (rt_movie_info['theater_date'] <= '2018-12-31')].copy()
movies_from_2000_2018
```

```python
#check the missing values again in movie_from_2000_2018 dataframe in each column
movies_from_2000_2018.isna().sum()
```

```python
#because the amount of NaN values in 'runtime' column is small compared to the amount of the dataset
#eliminating rows with NaN values in 'runtime' column
#check the NaN values again and see whether the 'runtime' column now has 0 NaN values
movies_from_2000_2018 = movies_from_2000_2018[pd.notnull(movies_from_2000_2018['runtime'])].copy()
movies_from_2000_2018.isna().sum()
```

```python
movies_from_2000_2018.shape
```

```python
print(movies_from_2000_2018['runtime'].dtype)
```

```python
movies_from_2000_2018['runtime'] = movies_from_2000_2018['runtime'].map(lambda x: x.split(' ', 1)[0]).copy()
```

```python
movies_from_2000_2018
```

```python
print(movies_from_2000_2018['runtime'].dtype)
```

```python
movies_from_2000_2018['runtime'] = movies_from_2000_2018['runtime'].astype(str).astype(int)
print(movies_from_2000_2018['runtime'].dtypes)
```
